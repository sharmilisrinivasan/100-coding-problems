```python
from collections import Counter
import heapq
```

```python
def get_top_k_freq_elements(list_of_elements, k):
    to_return = []

    if not list_of_elements or k<=0:
        return to_return
    
    # 1. counter to create num -> frequency => O(n)
    num_freq_map = Counter(list_of_elements)

    # 2. Create Heap to store (-freq, num) as python only stores min heap 
    #  => Time Complexity: `O(nlogn)` : log(n) to insert => For n elements = O(nlogn)
    #  => Space Complexity: O(n) for the heap
    max_heap = []

    for num, freq in num_freq_map.items():
        heapq.heappush(max_heap, (-freq, num))

    # 3. Keep poping top k from the heap
    for _ in range(k):
        if not max_heap:
            break
        neg_freq, num = heapq.heappop(max_heap)
        to_return.append(num)

    return to_return

def get_top_k_freq_elemeents_optimised(list_of_elements, k):
    to_return = []

    if not list_of_elements or k<=0:
        return to_return
    
    # 1. Counter to create num -> frequency => O(n)
    num_freq_map = Counter(list_of_elements)

    # 2. As the ask is only top "k", we can limit the heap size to "k"
    #  => Time Complexity: `O(nlogk)` : log(k) to insert => For n elements O(nlog(k))
    #  => Space Complaexity: O(k) for the k-heap
    max_heap = []

    for num, freq in num_freq_map.items():
        heapq.heappush(max_heap, (freq, num))  # Save as min heap to pop least freq

        if (len(max_heap) > k):
            heapq.heappop(max_heap)

    # 3. Return all elements in the k-heap
    for _,num in max_heap:
        to_return .append(num)

    return to_return

    
```

```python
# Test

# Good case
list_of_elements = [1,1,1,2,2,3]
k = 2

print(get_top_k_freq_elements(list_of_elements, k)) # Expected Output: [1,2]
print(get_top_k_freq_elemeents_optimised(list_of_elements, k)) # Expected Output: [1,2]
```

```python
# Empty list
list_of_elements = []
k = 2

print(get_top_k_freq_elements(list_of_elements, k)) # Expected Output: []
print(get_top_k_freq_elemeents_optimised(list_of_elements, k)) # Expected Output: []
```

```python
# k is negative
list_of_elements = [1,1,1,2,2,3]
k = -1
print(get_top_k_freq_elements(list_of_elements, k)) # Expected Output: []
print(get_top_k_freq_elemeents_optimised(list_of_elements, k)) # Expected Output: []
```

```python
# k is zero
list_of_elements = [1,1,1,2,2,3]
k = 0
print(get_top_k_freq_elements(list_of_elements, k)) # Expected Output: []
print(get_top_k_freq_elemeents_optimised(list_of_elements, k)) # Expected Output: []
```

```python
# k is more than unique elements
list_of_elements = [1,1,1,2,2,3]
k = 5
print(get_top_k_freq_elements(list_of_elements, k)) # Expected Output: [1,2,3]
print(get_top_k_freq_elemeents_optimised(list_of_elements, k)) # Expected Output: [1,2,3]
```
